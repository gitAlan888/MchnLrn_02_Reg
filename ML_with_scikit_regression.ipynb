```python
import pandas as pd
```

```python
auto_data = pd.read_csv('imports-85_data2.csv')
```

```python
import numpy as np
```

```python
auto_data = auto_data.replace('?', np.nan)
auto_data.head()
```

```python
auto_data.describe()
```

```python
auto_data.describe(include ='all') 
```

```python
auto_data['price'] = pd.to_numeric(auto_data['price'],errors='coerce')
```

```python
auto_data['price'].describe()
```

```python
auto_data = auto_data.drop('normalized-losses', axis=1)
auto_data.head()
```

```python
auto_data.describe()
```

```python
auto_data['horsepower'] = pd.to_numeric(auto_data['horsepower'],errors='coerce')
auto_data['horsepower'].describe()
```

```python
cylinders_dict={'two':2,
                'three':3,
                'four':4,
                'five':5,
                'six':6,
                'eight':8,
                'twelve':12}

#auto_data.describe(include='all')#.replace(cylinders_dict,inplace=True)

auto_data['num-of-cylinders'].replace(cylinders_dict,inplace=True)

```

```python
auto_data.head()
```

```python
auto_data['num-of-cylinders'].head()
```

```python
auto_data = pd.get_dummies(auto_data, columns=['make',
                                             'fuel-type',
                                             'aspiration',
                                             'num-of-doors',
                                             'body-style',
                                             'drive-wheels',
                                             'engine-location',
                                             'engine-type',
                                             'fuel-system'])
auto_data.head()
```

```python
auto_data = auto_data.dropna()
```

```python
auto_data[auto_data.isnull().any(axis=1)]
```

## FEED into MODEL

```python
from sklearn.model_selection import train_test_split

#X dataset to be used to train. Make sure to drop the target coloumn
X = auto_data.drop('price', axis=1)

# Y is the target column
Y = auto_data['price']

#Splitting into 80% for training set and 20% for testing set so we can see our accuracy
X_train,x_test, Y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()
linear_model.fit(X_train,Y_train)
```

```python
linear_model.score(X_train, Y_train)
```

```python
linear_model.coef_
```

```python
predictors = X_train.columns
coef = pd.Series(linear_model.coef_, predictors.sort_values())
print(coef)
```

```python
y_predict = linear_model.predict(x_test)
```

```python
%pylab inline 
#inline so it plots in here than opening a new window
pylab.rcParams['figure.figsize'] = (15,6)

plt.plot (y_predict, label='Predicted')
plt.plot(y_test.values, label='Actual')
plt.ylabel('price')

plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

linear_model_mse = mean_squared_error(y_predict, y_test)
linear_model_mse
```

```python
from sklearn.linear_model import Lasso

lasso_model = Lasso(alpha=10, normalize=True)
lasso_model.fit(X_train, Y_train)
```

```python
lasso_model.score(X_train, Y_train)
```

```python
predictors = X_train.columns
coef = pd.Series(lasso_model.coef_, predictors).sort_values()
print(coef)
```

```python
import math
from sklearn.metrics import mean_squared_error
y_predict = lasso_model.predict(x_test)

%pylab inline 
#inline so it plots in here than opening a new window
pylab.rcParams['figure.figsize'] = (15,6)

plt.plot (y_predict, label='Predicted')
plt.plot(y_test.values, label='Actual')
plt.ylabel('price')

plt.legend()
plt.show()

r_square = lasso_model.score(x_test,y_test)


lasso_model_mse = mean_squared_error(y_predict, y_test)
lasso_model_mse

print('R-Square : ', r_square)
print('Math Sqaure Root MSE : ', math.sqrt(lasso_model_mse))
```

```python
from sklearn.linear_model import Ridge

ridge_model = Ridge(alpha=0.5, normalize=True)
ridge_model.fit(X_train, Y_train)
```

```python
predictors = X_train.columns
coef = pd.Series(ridge_model.coef_, predictors).sort_values()
print(coef)
```

```python
ridge_model.score(X_train,Y_train)
```

```python
#import lib: Math and MSE
import math
from sklearn.metrics import mean_squared_error
#prediction on the test data set
y_predict = ridge_model.predict(x_test)


#plot the predicted vs actual chart
%pylab inline 
#inline so it plots in here than opening a new window
pylab.rcParams['figure.figsize'] = (15,6)

plt.plot (y_predict, label='Predicted')
plt.plot(y_test.values, label='Actual')
plt.ylabel('price')
plt.legend()
plt.show()

#Calculate predicted test vs actual test : Score (R -Square), MSE and RMSE
r_square = ridge_model.score(x_test,y_test)
ridge_model_mse = mean_squared_error(y_predict, y_test)

print('R-Square : ', r_square)
print('Mean Square Error (MSE) : ', ridge_model_mse)
print('Math Sqaure Root MSE (RMSE): ', math.sqrt(ridge_model_mse))
```
